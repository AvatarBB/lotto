```python
import numpy as np
```

```python
gap = 30
# set lookback value 
def combination(data):
    dataset = []
    for idx,value in enumerate(data):
        if(idx<len(data)-gap):
            temp = data[idx:idx+gap]
            temp = temp.reshape(6*gap)
            dataset.append(temp)
    return np.array(dataset)

def get_data():#normlize
    data = np.loadtxt('hk.txt')
    print(data.shape)
    x = combination(data)
    y = data[gap:]
    return x,y

x,y = get_data()
```

```python
# Deep NN with keras
from keras.models import Sequential 
from keras.layers.core import Dense, Dropout
from keras.layers import Embedding,LSTM
from keras import backend as K

def swish(x):   #Activation Function
    return x*K.sigmoid(x)

def get_LSTM():
    model = Sequential()
    model.add(Embedding(6*gap,32))
    model.add(LSTM(512,recurrent_dropout=0.2, dropout=0.2))
    model.add(Dropout(0.2))
    model.add(Dense(256, activation='tanh'))
    model.add(Dropout(0.5))
    model.add(Dense(64, activation='tanh'))
    model.add(Dropout(0.2))
    model.add(Dense(6, activation="sigmoid")) 
    model.summary()
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model        
```

```python
def get_test():
    data = np.loadtxt('lotto.txt').astype('int8')
    print(data.shape)
    x = combination(data)
    y = data[gap:]
    return x,y 

test_x,test_y = get_test()
```

```python
lstm_model = get_LSTM()
lstm_model.fit(x/50, y/50, batch_size=32, epochs=1,verbose=1)
lstm_result =np.around(lstm_model.predict(test_x/50)*50).astype('int8')
```

```python
lstm_model.save('model.h5')
```
