```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
def get_data():
    data = np.loadtxt('hk.txt').astype('int8')
    print(data.shape)
    return data
```

```python
data = get_data()
print(data)
```

```python
flat = data.flatten().tolist()
element = {k:flat.count(k) for k in set(flat)}
print(element)
plt.figure(figsize=(20,7)) 
plt.bar(range(len(element)), list(element.values()), align='center',color='y')
plt.xticks(range(len(element)), list(element.keys()))
plt.show()
```

```python
maxElement = element[max(element, key=element.get)]
minElement = element[min(element, key=element.get)]
print(max(element, key=element.get),'appear',maxElement,'times')
print(min(element, key=element.get),'appear',minElement,'times')
rate = (maxElement-minElement)/maxElement
print(round(rate,5)*100,'% higher than min appears')
```
